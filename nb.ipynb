```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('lena.jpg').convert('L')
img
```

```python
# Convert image to array in uint8
img_array = np.array(img, dtype=np.uint8)
img_array
```

```python
img_array.shape
```

```python
# Generate a random array
random_array = np.random.randint(0, 255, size=(225, 400))
random_array
```

```python
plt.imshow(random_array, cmap='gray')  # Display the array as a grayscale image
plt.show()
```

```python
plt.imshow(img_array, cmap='gray')  # Display the array as a grayscale image
plt.show()
```

```python
# Find all values greater than 255
values_greater_than_255 = img_array[img_array > 255]
print("Values greater than 255:", values_greater_than_255)
# Find all values equal to 255
values_equal_to_255 = img_array[img_array == 255]
print("Values equal to 255:", values_equal_to_255)
# Find all values equal to 0
values_equal_to_0 = img_array[img_array == 0]
print("Values equal to 0:", values_equal_to_0)
```

```python
# Aumentar el brillo
img_array_brighter = img_array + 30  # Aumentar el valor de cada pxel en 30
img_array_brighter
```

### saturating arithmetic
Saturating arithmetic means that if the result of an operation exceeds the maximum value representable by the data type (255 in the case of uint8), the result is clamped to the maximum value instead of overflowing.

```python
# Find all values greater than 255
values_greater_than_255 = img_array_brighter[img_array_brighter > 255]
print("Values greater than 255:", values_greater_than_255)
# Find all values equal to 255
values_equal_to_255 = img_array_brighter[img_array_brighter == 255]
print("Values equal to 255:", values_equal_to_255)
# Find all values equal to 0
values_equal_to_0 = img_array_brighter[img_array_brighter == 0]
print("Values equal to 0:", values_equal_to_0)
```

```python
plt.imshow(img_array_brighter, cmap='gray')  # Display the array as a grayscale image
plt.show()
```
